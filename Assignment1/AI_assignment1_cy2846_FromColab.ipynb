```python
import torch
import matplotlib.pyplot as plt

true_lambda = 100.0       # Arrival rate
N = 1000

# Set random seed for reproducibility
torch.manual_seed(5)

# Simulate interarrival times using an exponential distribution
exp_dist = torch.distributions.Exponential(rate=true_lambda)
interarrival_times = exp_dist.sample((N,))

# Generate x values for the theoretical PDF plot
x = torch.linspace(0, interarrival_times.max(), steps=1000)
# Compute the theoretical PDF: f(x) = lambda * exp(-lambda * x)
pdf = true_lambda * torch.exp(-true_lambda * x)

# Plot the theoretical PDF
plt.figure(figsize=(8, 5))
plt.plot(x.numpy(), pdf.numpy(), 'r-', lw=2, label='Theoretical PDF')

# Plot the histogram of simulated interarrival times (normalized to form a density)
plt.hist(interarrival_times.numpy(), bins=30, density=True, alpha=0.6, color='b')
plt.xlabel("Interarrival Time (seconds)")
plt.ylabel("Density")
plt.title("Histogram of Simulated Interarrival Times")
plt.legend()
plt.show()
```

```python
import torch
import matplotlib.pyplot as plt

true_lambda = 100.0            # Arrival rate lambda
N = 1000

# Set random seed for reproducibility
torch.manual_seed(5)

# Simulate data from an exponential distribution (scale = 1 / lambda)
exp_dist = torch.distributions.Exponential(rate=true_lambda)
data = exp_dist.sample((N,))

# SGD settings
learning_rate = 0.01
num_iterations = 2000000
lambda_est = torch.tensor(20.0)  # Initial guess for lambda (1/0.05 = 20)

objective_values = []
iteration_indices = []

def full_objective(lmbd, data):
    # L() = -N * log() +  * sum(data)
    return -len(data) * torch.log(lmbd) + lmbd * torch.sum(data)

# SGD Loop
for iteration in range(num_iterations):
    # Randomly pick an index for the current data sample
    i = torch.randint(low=0, high=N, size=(1,)).item()
    x_i = data[i]

    # Compute the gradient: dL/d = -1/ + x_i
    grad = -1.0 / lambda_est + x_i

    # Update the estimate of lambda using the gradient
    lambda_est = lambda_est - learning_rate * grad

    # Every 100 iterations, compute and record the full objective function value
    if iteration % 100 == 0:
        obj_val = full_objective(lambda_est, data)
        objective_values.append(obj_val.item())
        iteration_indices.append(iteration)

print("Estimated lambda after SGD:", lambda_est.item())

# Compute the analytical MLE for lambda: lambda_hat = N / sum(data)
lambda_MLE = N / torch.sum(data).item()
print("Analytical MLE for lambda:", lambda_MLE)

# Plot the objective function value versus iteration
plt.figure(figsize=(8, 5))
plt.plot(iteration_indices, objective_values, linestyle='-', color='b')
plt.xlabel("Iteration")
plt.ylabel("Negative Log-Likelihood")
plt.title("Objective Function Value During SGD")
plt.grid(True)
plt.show()
```

##Linear Regression and SGD

```python
import torch
import matplotlib.pyplot as plt

# Function to generate toy dataset using PyTorch
def create_toy_data(func, sample_size, std, domain=(0, 1)):
    x = torch.linspace(domain[0], domain[1], sample_size)
    indices = torch.randperm(sample_size)
    x = x[indices]
    y = func(x) + torch.normal(mean=0.0, std=std, size=(sample_size,))
    return x, y

def sinusoidal(x):
    return torch.sin(2 * torch.pi * x)

# Define a function to construct polynomial features
def polynomial_features(x, degree):
    features = torch.stack([x**i for i in range(degree+1)], dim=1)
    return features

# Generate training dataset
sample_size = 500
x_train, y_train = create_toy_data(sinusoidal, sample_size, std=0.25)

# Generate polynomial features (here using degree = 8)
degree = 8
X_train = polynomial_features(x_train, degree)  # X_train shape: (500, degree+1)

# Set hyperparameters: learning rate, number of epochs, L2 regularization coefficient
learning_rate = 0.01
num_epochs = 300
lambda_reg = 0.01            # L2 regularization coefficient

# Initialize model parameters w randomly (degree+1 parameters)
w = torch.randn(degree + 1)

loss_history = []
m = x_train.shape[0]

# Train the model using SGD with L2 regularization
for epoch in range(num_epochs):
    # Shuffle the dataset at the beginning of each epoch
    indices = torch.randperm(m)
    for idx in indices:
        xi = X_train[idx]
        yi = y_train[idx]

        # Compute prediction: dot product of xi and weights w
        pred = torch.dot(xi, w)
        error = pred - yi

        # Compute gradient: MSE gradient for one sample plus L2 regularization term
        gradient = xi * error + lambda_reg * w

        # Update weights using the gradient
        w = w - learning_rate * gradient

    # After each epoch, compute the loss over the full dataset
    predictions = X_train @ w
    errors = predictions - y_train
    loss = (errors.pow(2).mean() + lambda_reg * w.pow(2).sum()) / 2
    loss_history.append(loss.item())

# Plot loss versus epoch to observe convergence
plt.figure(figsize=(8, 5))
plt.plot(loss_history, label="Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("SGD Convergence with Polynomial Features")
plt.legend()
plt.grid(True)
plt.show()

#  Plot training data and the model's prediction curve
x_range = torch.linspace(torch.min(x_train), torch.max(x_train), steps=200)
X_range = polynomial_features(x_range, degree)
y_pred = X_range @ w  # Compute predictions

plt.figure(figsize=(8, 5))
plt.scatter(x_train.numpy(), y_train.numpy(), color='blue', label="Training Data")
plt.plot(x_range.numpy(), y_pred.numpy(), color='red', label="Prediction Curve", linewidth=2)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Multivariate Linear Regression Fitting Sinusoidal Data")
plt.legend()
plt.grid(True)
plt.show()

# Print the hyper-parameters and final weights
print("Hyper-parameters:")
print("  Learning Rate =", learning_rate)
print("  Number of Epochs =", num_epochs)
print("  Polynomial degree =", degree)
print("  L2 Regularization (lambda) =", lambda_reg)
print("  Sample Size =", sample_size)
print("\nFinal estimated weights (w):", w)
```

According to the resulting figure of SGD loss curve, the algorithm has been well converged. Measures to prevent overfitting also work well. Combined with the prediction curve in the second figure, it shows that the algorithm is good and effective.
